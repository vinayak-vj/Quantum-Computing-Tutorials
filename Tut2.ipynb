```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from qiskit_aer import Aer
```

```python
from qiskit import QuantumCircuit
qc1=QuantumCircuit(1,1)
qc1.h(0)
qc1.draw('mpl')
```

```python
qc1.measure(range(1),range(1))
qc1.draw('mpl')
```

```python
# Use local simulator
aer_sim = Aer.get_backend('aer_simulator')
shots = 4096                                  # The number of times a quantum circuit is executed (or simulated) to gather measurement data.
results = aer_sim.run(qc1, shots=shots).result()
sim_counts = results.get_counts()             # Counts are a dictionary-like output that records how many times each possible measurement outcome occurred during the execution of the quantum circuit.

print(sim_counts)
plot_histogram(sim_counts)
```

Q5

```python
qc5=QuantumCircuit(3,3)
qc5.ccx(0,1,2)
qc5.x(0)
qc5.h(2)
qc5.cx(0,1)
qc5.draw('mpl')
```

```python
qc5.measure(range(3),range(3))
qc5.draw('mpl')
```

```python
aer_sim=Aer.get_backend('aer_simulator')
shots=[10,50,100,500,1000]
results=[]
for i in shots:
  result=aer_sim.run(qc5,shots=i).result()
  counts=result.get_counts()
  results.append(counts)

print(results)
plot_histogram(results)
```

```python
run_numbers = [i + 1 for i in range(len(results))]

# Extract counts for '111' and '011' from the results
count_111 = [result.get('111', 0) for result in results]
count_011 = [result.get('011', 0) for result in results]


plt.scatter(run_numbers, count_111, color='blue', label="Count of '111'", marker='o')


plt.scatter(run_numbers, count_011, color='red', label="Count of '011'", marker='x')


plt.title('Measurement Results Across Runs')
plt.xlabel('Run Number')
plt.ylabel('Count')
plt.legend()

plt.grid(True)
plt.show()
```
